```python
import scipy.io
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=scipy.io.loadmat('/mnt/downloads/_aMultipower_ZSrsq.mat')
dataset=dataset['ZS2_rsq']
```

```python
dataset.shape
```

```python
import hdbscan
```

```python
hdbscan.dist_metrics.METRIC_MAPPING
```

```python
clusterer = hdbscan.HDBSCAN(metric='manhattan',min_cluster_size=50, min_samples=20).fit(dataset)
```

```python
from sklearn.manifold import TSNE
```

```python
projection = TSNE(n_components=2, perplexity=50.0).fit_transform(dataset)
```

```python
plt.figure(figsize = (30,30))
plt.scatter(*projection.T)
```

```python
import seaborn as sns
```

```python
hdbscan.hdbscan_.DistanceMetric
```

```python
color_palette = sns.color_palette('Paired', 12)
cluster_colors = [color_palette[x] if x >= 0
                  else (0.5, 0.5, 0.5)
                  for x in clusterer.labels_]
cluster_member_colors = [sns.desaturate(x, p) for x, p in
                         zip(cluster_colors, clusterer.probabilities_)]
```

```python
plt.figure(figsize = (30,30))
plt.scatter(*projection.T, s=50, linewidth=0, c=cluster_member_colors, alpha=0.25)
```
