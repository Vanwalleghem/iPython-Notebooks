```python
import skimage
import os
import glob
import numpy as np
```

```python
glob.glob(os.path.join('d:/Pictures/processed/Itia/MultiPower/','*_mean.tif'))
```

```python
fnames_all=[]
for file in glob.glob(os.path.join('d:/Pictures/processed/Itia/MultiPower/','*_mean.tif')):
    if file.endswith(".tif") and not file.endswith("_mean.tif"):
        fnames_all.append(file)
FishNames=list(set([s.partition("Fish")[2].partition('_')[0] for s in fnames_all]))
```

```python
ELO_list=[]
ILO_list=[]
CLO_list=[]
OLO_list=[]
for FishName in FishNames:
    ELO_list.append([s for s in fnames_all if ('ELO' in s.partition("AVG_")[2]) & (FishName in s)])
    ILO_list.append([s for s in fnames_all if ('ILO' in s.partition("AVG_")[2]) & (FishName in s)])
    CLO_list.append([s for s in fnames_all if ('CLO' in s.partition("AVG_")[2]) & (FishName in s)])
    OLO_list.append([s for s in fnames_all if ('OLO' in s.partition("AVG_")[2]) & (FishName in s)])
```
